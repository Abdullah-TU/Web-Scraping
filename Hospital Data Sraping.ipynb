# Hosepital Data

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

url = "https://www.bkk-klinikfinder.de/suche/suchergebnis.php?next=1"

# Set up the Selenium webdriver
driver = webdriver.Chrome()

# Open the URL
driver.get(url)

# Wait for the page to load (you may need to adjust the time based on the website)
driver.implicitly_wait(20)

# Define a loop to click the "More" button and extract information
counter = 1
while True:
    try:
        # Execute JavaScript to click the "More" button
        driver.execute_script("arguments[0].click();", driver.find_element(By.ID, 'more_btn'))

        # Wait for the newly loaded content to appear
        time.sleep(5)  # Adjust the delay as needed

        # Find all the <a> tags with class "main-link"
        elements = driver.find_elements(By.CLASS_NAME, 'main-link')

        # Iterate through the elements and print the text and links
        for element in elements:
            href = element.get_attribute('href')
            text = element.text

            if href:
                print(f'{counter}: {text}\nLink: {href}\n')
                counter += 1

    except Exception as e:
        print(f"An error occurred: {e}")
        break

# Close the browser
driver.quit()
```

# Saving data in a CSV file


### Total number of links and text: Unknown. Downloaded more than 226,543 links and text.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv

url = "https://www.bkk-klinikfinder.de/suche/suchergebnis.php?next=1"

# Set up the Selenium webdriver
driver = webdriver.Chrome()

# Open the URL
driver.get(url)

# Wait for the page to load (you may need to adjust the time based on the website)
driver.implicitly_wait(20)

# Define a loop to click the "More" button and extract information
counter = 1

# Create a CSV file and write headers
with open('hospital_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Number', 'Text', 'Link'])

    while True:
        try:
            # Execute JavaScript to click the "More" button
            driver.execute_script("arguments[0].click();", driver.find_element(By.ID, 'more_btn'))

            # Wait for the newly loaded content to appear
            driver.implicitly_wait(10)

            # Find all the <a> tags with class "main-link"
            elements = driver.find_elements(By.CLASS_NAME, 'main-link')

            # Iterate through the elements and write to CSV
            for element in elements:
                href = element.get_attribute('href')
                text = element.text

                if href:
                    writer.writerow([counter, text, href])
                    counter += 1

        except Exception as e:
            if "no such window: target window already closed" in str(e):
                print("Window closed. Exiting loop.")
            else:
                print(f"An error occurred: {e}")
            break

# Close the browser
driver.quit()
```
